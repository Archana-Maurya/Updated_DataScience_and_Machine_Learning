```python
# Install required libraries
!pip install -q google-generativeai transformers

# Import libraries
import google.generativeai as genai
from transformers import AutoTokenizer

# Configure Gemini API Key
# Replace YOUR_NEW_API_KEY with your new Gemini API key
genai.configure(api_key="AIzaSyAQ2E1YcLZvSlCQpII8rR9OIyXJvMPbfGo")

# Load Gemini model
model = genai.GenerativeModel("gemini-flash-latest")

# Sample research article
article = """
Artificial Intelligence (AI) is transforming many industries including healthcare,
education, finance, and transportation. Machine learning algorithms allow computers
to learn patterns from large datasets and make predictions or decisions without
explicit programming.

In healthcare, AI systems assist doctors by analyzing medical images, predicting
disease risks, and recommending treatments. In education, AI-driven platforms
personalize learning experiences based on student performance.

However, AI also introduces challenges such as ethical concerns, bias in algorithms,
and data privacy issues. Researchers are actively exploring methods to build
transparent and responsible AI systems.

The future of AI will likely involve collaboration between humans and intelligent
machines, improving productivity and innovation across sectors.
"""

# ---------------- ZERO SHOT PROMPT ----------------
zero_prompt = f"""
Summarize the following article in 5 bullet points:

{article}
"""

zero_result = model.generate_content(zero_prompt)

print("\n===== ZERO SHOT SUMMARY =====\n")
print(zero_result.text)


# ---------------- FEW SHOT PROMPT ----------------
few_prompt = f"""
Example 1
Article: Renewable energy sources like solar and wind reduce dependence on fossil fuels.
Summary:
• Renewable energy reduces carbon emissions
• Solar and wind are major sources
• Helps fight climate change

Example 2
Article: Cloud computing allows businesses to store and process data remotely.
Summary:
• Data storage happens on remote servers
• Businesses save infrastructure cost
• Improves scalability

Now summarize the following article in the same format:

{article}
"""

few_result = model.generate_content(few_prompt)

print("\n===== FEW SHOT SUMMARY =====\n")
print(few_result.text)


# ---------------- CHAIN OF THOUGHT PROMPT ----------------
cot_prompt = f"""
Analyze the article step by step:

1. Identify the main topic
2. Extract key ideas
3. Generate a concise summary

Article:
{article}
"""

cot_result = model.generate_content(cot_prompt)

print("\n===== CHAIN OF THOUGHT ANALYSIS =====\n")
print(cot_result.text)


# ---------------- EXECUTIVE SUMMARY PROMPT ----------------
opt_prompt = f"""
You are an academic research assistant.

Analyze the article and produce a professional executive summary.

Output format:

Key Insights:
1.
2.
3.

Actionable Takeaway:
1 practical recommendation based on the article.

Article:
{article}
"""

opt_result = model.generate_content(opt_prompt)

print("\n===== EXECUTIVE SUMMARY =====\n")
print(opt_result.text)


# ---------------- TOKENIZATION WITH BERT ----------------
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

text = """
Artificial Intelligence helps computers learn patterns from data
and make predictions without explicit programming.
"""

tokens = tokenizer.tokenize(text)

print("\n===== TOKENS =====")
print(tokens)

print("\nTotal Tokens:", len(tokens))


# ---------------- RESEARCH ASSISTANT FUNCTION ----------------
def research_assistant(article):

    prompt = f"""
    You are an AI research assistant.

    Analyze the article and return:

    1. Short Summary
    2. Key Insights (3 points)
    3. One Actionable Recommendation

    Article:
    {article}
    """

    response = model.generate_content(prompt)

    return response.text


print("\n===== AI RESEARCH ASSISTANT RESULT =====\n")
print(research_assistant(article))
```
